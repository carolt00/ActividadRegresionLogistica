```python
#Punto 4
#Porcentaje de datos para entrenamiento (train) = 75% y (test) = 25%
#Se crean variables X y Y para almacenar la informacin del entrenamiento

X = Datos.drop('Condicion',axis=1) #Se cargan los datos de train sin los datos de la columna Condicion
Y = Datos['Condicion'] #Se cargan los datos de train con los datos de la id columna Diabetico

#La funcion train_test_split crea una division de un conjunto de datos en dos bloques uno de entrenamiento y otro de prueba (train and test)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, stratify=Y, random_state=0)
```

```python
print(X_train)
```

```python
print(Y_train)
```

```python
#Punto 5
#Construir el modelo de regresin logistico

modelo_regresion = LogisticRegression(solver = "liblinear") #Libreria para problemas lineales de clasificacion y regresion
modelo_regresion.fit (X_train, Y_train)

#Crea el modelo de prediccion
Prediccion = modelo_regresion.predict(X_test) #Predice valores a partir del conjunto de datos de validacion
```

```python
print(Y_test)
```

```python
print(Prediccion)
```

```python
#Punto 6
#Desempeo del modelo de regresion

print(classification_report(Y_test,Prediccion))
```

```python
#Punto 7
#Matrix de confusion del modelo

cm=confusion_matrix(Y_test,Prediccion)
disp=metrics.ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
```

```python
#Curva de prediccion del modelo
from sklearn.metrics import RocCurveDisplay
RocCurveDisplay.from_predictions(Y_test, Prediccion)
plt.show()
```
